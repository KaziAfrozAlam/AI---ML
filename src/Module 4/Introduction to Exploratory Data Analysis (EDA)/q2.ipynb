**Title**: Understanding the Dataset <br>

**Task**: For the given datasets below, identify the data types and dimensions.<br>

Task 1: Employee Dataset<br>
Columns: Employee ID , Name , Age , Department , Salary , Joining Date<br>
Task 2: Product Sales Dataset<br>
Columns: Product ID , Product Name , Price , Quantity Sold , Category , Sales Date<br>
Task 3: Student Grades Dataset<br>
Columns: Student ID , Student Name , Math Score , Science Score , English Score ,Year<br>

Instructions:<br>

Identify which columns are numerical (continuous or discrete) and which are categorical
(nominal or ordinal).<br>
Note down the dimensions (number of rows and columns) of the dataset.

```python
# Write your code from here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

employee_df = pd.DataFrame({
    'Employee ID': range(1, 11),
    'Name': ['Emp'+str(i) for i in range(1, 11)],
    'Age': [25, 30, 45, 28, 33, 41, 60, 29, 31, 120],  
    'Department': ['HR', 'Sales', 'Tech', 'HR', 'Tech', 'Sales', 'Tech', 'HR', 'Sales', 'HR'],
    'Salary': [40000, 50000, 120000, 45000, 70000, 80000, 95000, 47000, 51000, 100000],
    'Joining Date': pd.date_range(start='2015-01-01', periods=10, freq='Y')
})

sales_df = pd.DataFrame({
    'Product ID': range(101, 111),
    'Product Name': ['Product'+str(i) for i in range(1, 11)],
    'Price': [20, 25, 23, 500, 24, 22, 21, 27, 20, 23],  # 500 is an outlier
    'Quantity Sold': [100, 150, 120, 10, 130, 140, 160, 180, 110, 125],
    'Category': ['A', 'A', 'B', 'C', 'B', 'A', 'B', 'C', 'A', 'C'],
    'Sales Date': pd.date_range(start='2021-01-01', periods=10, freq='M')
})

grades_df = pd.DataFrame({
    'Student ID': range(201, 211),
    'Student Name': ['Student'+str(i) for i in range(1, 11)],
    'Math Score': [75, 80, 60, 90, 88, 95, 40, 65, 100, 5],  
    'Science Score': [78, 82, 61, 91, 89, 96, 45, 67, 98, 10],
    'English Score': [72, 85, 66, 88, 83, 94, 50, 70, 99, 60],
    'Year': ['2020', '2020', '2021', '2021', '2022', '2022', '2020', '2021', '2022', '2021']
})
```

**Title**: Checking for Missing Values<br>

**Task**: Identify and count the number of missing values in each dataset.<br>

Instructions:<br>
Use Python or any data manipulation tool to check for missing values in each column of the datasets. <br>Report the columns which have missing values and their counts.

```python
# Write your code from here
print("Employee Dataset Info:")
print(employee_df.dtypes)
print("Shape:", employee_df.shape)

print("\nProduct Sales Dataset Info:")
print(sales_df.dtypes)
print("Shape:", sales_df.shape)

print("\nStudent Grades Dataset Info:")
print(grades_df.dtypes)
print("Shape:", grades_df.shape)

print("\nMissing Values in Employee Dataset:\n", employee_df.isnull().sum())
print("\nMissing Values in Product Sales Dataset:\n", sales_df.isnull().sum())
print("\nMissing Values in Student Grades Dataset:\n", grades_df.isnull().sum())
```

**Title**: Handling Outliers<br>

**Task**: Detect and propose handling methods for outliers in the numerical columns of the datasets.<br>

Task 1: Age in Employee Dataset<br>
Task 2: Price in Product Sales Dataset<br>
Task 3: Math Score in Student Grades Dataset<br>

Instructions:<br>

Use box plots to visualize potential outliers.<br>
Suggest methods to handle them, such as removal or transformation.

```python
# Write your code from here
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x=employee_df['Age'])
plt.title("Outliers in Age (Employee)")

plt.subplot(1, 3, 2)
sns.boxplot(x=sales_df['Price'])
plt.title("Outliers in Price (Sales)")

plt.subplot(1, 3, 3)
sns.boxplot(x=grades_df['Math Score'])
plt.title("Outliers in Math Score (Grades)")
plt.tight_layout()
plt.show()
```

**Title**: Finding Relationships Between Features<br>

**Task**: Identify relationships between pairs of features in the datasets.<br>

Task 1: Salary vs Age in Employee Dataset<br>
Task 2: Price vs Quantity Sold in Product Sales Dataset<br>
Task 3: Math Score vs Science Score in Student Grades Dataset

Instructions: <br>

Use scatter plots or correlation coefficients to analyze the relationships.<br>
Describe any insights or patterns observed.

```python
# Write your code from here
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.scatterplot(x=employee_df['Age'], y=employee_df['Salary'])
plt.title("Age vs Salary")

plt.subplot(1, 3, 2)
sns.scatterplot(x=sales_df['Price'], y=sales_df['Quantity Sold'])
plt.title("Price vs Quantity Sold")

plt.subplot(1, 3, 3)
sns.scatterplot(x=grades_df['Math Score'], y=grades_df['Science Score'])
plt.title("Math Score vs Science Score")
plt.tight_layout()
plt.show()

print("\nCorrelation (Age vs Salary):", employee_df[['Age', 'Salary']].corr().iloc[0, 1])
print("Correlation (Price vs Quantity):", sales_df[['Price', 'Quantity Sold']].corr().iloc[0, 1])
print("Correlation (Math vs Science):", grades_df[['Math Score', 'Science Score']].corr().iloc[0, 1])
```
