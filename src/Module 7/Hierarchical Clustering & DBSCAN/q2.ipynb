```python
# Question 2: Understanding Dendrograms
# Description: Generate a dendrogram to visualize the clustering process.

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
X = np.array([
    [1, 2],
    [2, 3],
    [3, 1],
    [6, 7],
    [7, 6],
    [8, 8]
])
Z = linkage(X, method='ward')

plt.figure(figsize=(8, 5))
dendrogram(Z, leaf_rotation=90, leaf_font_size=10)
plt.title("Dendrogram: Visualizing the Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance (Euclidean)")
plt.grid(True)
plt.show()
```
