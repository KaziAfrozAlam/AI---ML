```python

#Title:  DBSCAN Clustering for Anomaly Detection

# Task 1: DBSCAN on Uniform Data

# 1. Generate a dataset with points [(-1, -1), (0, 0), (1, 1), (9, 9), (10, 10)] .
# 2. Apply DBSCAN with a small eps (e.g., 0.5) and a minimum number of points (e.g., 2).
# 3. Identify which points are treated as noise/outliers.

import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Task 1: DBSCAN on Uniform Data
data1 = np.array([[-1, -1], [0, 0], [1, 1], [9, 9], [10, 10]])
db1 = DBSCAN(eps=0.5, min_samples=2).fit(data1)
labels1 = db1.labels_

print("Task 1: DBSCAN on Uniform Data")
print("Labels:", labels1)
print("Outliers:", data1[labels1 == -1].tolist())

plt.figure(figsize=(6, 4))
plt.scatter(data1[:, 0], data1[:, 1], c=labels1, cmap='Set1', s=100, edgecolors='k')
plt.title("Task 1: DBSCAN (eps=0.5, min_samples=2)")
plt.grid(True)
plt.tight_layout()
plt.show()


# Task 2: Detecting Anomalies in a Simple Dataset

# 1. Use the dataset: [(0, 0), (0, 1), (1, 0), (10, 10), (11, 11)] .
# 2. Configure DBSCAN parameters ( eps=1 , min_samples=2 ) and apply it.
# 3. Analyze the resulting clusters and classify any points identified as outliers.


# Task 2: Detecting Anomalies in a Simple Dataset
data2 = np.array([(0, 0), (0, 1), (1, 0), (10, 10), (11, 11)])
db2 = DBSCAN(eps=1, min_samples=2).fit(data2)
labels2 = db2.labels_

print("\nTask 2: Detecting Anomalies")
print("Labels:", labels2)
print("Outliers:", data2[labels2 == -1].tolist())

plt.figure(figsize=(6, 4))
plt.scatter(data2[:, 0], data2[:, 1], c=labels2, cmap='Set2', s=100, edgecolors='k')
plt.title("Task 2: DBSCAN (eps=1, min_samples=2)")
plt.grid(True)
plt.tight_layout()
plt.show()


# Task 3: Exploring Real World Data

# 1. Load a dataset with anomalies like credit card transactions.
# 2. Apply DBSCAN to detect unusual transactions by adjusting eps and min_samples .
# 3. Identify and discuss the points that are marked as noise, understanding why they might be considered anomalies.

X, _ = make_blobs(n_samples=100, centers=1, cluster_std=0.5, random_state=42)
anomalies = np.array([[8, 8], [9, 9]])
data3 = np.vstack([X, anomalies])

db3 = DBSCAN(eps=0.5, min_samples=5).fit(data3)
labels3 = db3.labels_

print("\nTask 3: Real-World-like Anomaly Detection")
print("Outliers:", data3[labels3 == -1].tolist())

plt.figure(figsize=(6, 4))
plt.scatter(data3[:, 0], data3[:, 1], c=labels3, cmap='Accent', s=50, edgecolors='k')
plt.title("Task 3: Simulated Real-World DBSCAN")
plt.grid(True)
plt.tight_layout()
plt.show()
```
