```python
# Question 5: Modify a hierarchical clustering algorithm to perform clustering using a custom distance metric on 2D points.

import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

points = np.array([
    [1, 2],
    [3, 4],
    [5, 1],
    [6, 3],
    [7, 2]
])

def manhattan_distance(u, v):
    return np.sum(np.abs(u - v))

distance_matrix = pdist(points, metric=manhattan_distance)
Z = linkage(distance_matrix, method='average')

plt.figure(figsize=(8, 4))
dendrogram(Z, labels=[f'P{i+1}' for i in range(len(points))])
plt.title('Hierarchical Clustering with Manhattan Distance')
plt.xlabel('Points')
plt.ylabel('Distance')
plt.show()
```
