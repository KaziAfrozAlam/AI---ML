```python
# Question 6: Implement an agglomerative hierarchical clustering on a non-Euclidian space (e.g.,textual data) using Gower's distance.

import numpy as np
import pandas as pd
import gower
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Age': [25, 30, 45, 35, 22],
    'Income': [50000, 60000, 80000, 56000, 45000],
    'Gender': ['M', 'F', 'F', 'M', 'F'],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
})

gower_dist = gower.gower_matrix(data)
condensed_dist = squareform(gower_dist, checks=False)

Z = linkage(condensed_dist, method='average')

plt.figure(figsize=(8, 4))
dendrogram(Z, labels=[f'Person{i+1}' for i in range(len(data))])

plt.title('Hierarchical Clustering with Gower Distance')
plt.xlabel('Individuals')
plt.ylabel('Distance')
plt.show()

```
