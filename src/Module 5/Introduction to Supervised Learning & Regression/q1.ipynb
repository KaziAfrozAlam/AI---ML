Question 1: Implementing Linear Regression <br>

Description:<br>
Implement a simple linear regression model using Python and scikit-learn to predict house prices based on one feature: the size of the house.

```python
# Write your code from here
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

house_size = np.array([650, 700, 720, 750, 800, 850, 900, 950, 1000, 1100]).reshape(-1, 1)
house_price = np.array([150, 155, 160, 165, 170, 175, 180, 190, 195, 210])

X_train, X_test, y_train, y_test = train_test_split(house_size, house_price, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared Score:", r2_score(y_test, y_pred))

plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($1000s)")
plt.title("Linear Regression - House Price Prediction")
plt.legend()
plt.show()
```
