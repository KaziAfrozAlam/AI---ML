Polynomial Regression

Task 1: Apply polynomial regression to a dataset showing the relationship between age and income.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)
age = np.random.randint(18, 70, size=100)
income = 20000 + age * 1000 + (age ** 2) * 150  

X = age.reshape(-1, 1)
y = income

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)

plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(np.sort(X, axis=0), poly_regressor.predict(poly.transform(np.sort(X, axis=0).reshape(-1, 1))), color='red', label='Polynomial Regression')
plt.title("Polynomial Regression: Age vs Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
plt.show()

```

Task 2: Explore and visualize a higher-degree polynomial fitting compared to a simple linear regression.

```python
# Write your code here

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
poly_10 = PolynomialFeatures(degree=10)
X_poly_10 = poly_10.fit_transform(X)
poly_regressor_10 = LinearRegression()
poly_regressor_10.fit(X_poly_10, y)

plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(np.sort(X, axis=0), linear_regressor.predict(np.sort(X, axis=0).reshape(-1, 1)), color='green', label='Linear Regression')
plt.plot(np.sort(X, axis=0), poly_regressor_10.predict(poly_10.transform(np.sort(X, axis=0).reshape(-1, 1))), color='red', label='Polynomial Regression (Degree 10)')
plt.title("Linear vs Polynomial Regression (Degree 10)")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
plt.show()

```

Task 3: Dataset with cyclical patterns such as tides or circadian cycles.

```python
# Write your code here
t = np.linspace(0, 24, 100)  
amplitude = 10  
offset = 50 
y_cyclical = amplitude * np.sin(t) + offset + np.random.normal(0, 2, size=len(t))  # Adding noise

plt.plot(t, y_cyclical, color='purple', label='Cyclical Data')
plt.title("Cyclical Data (Tides/Circadian Cycles)")
plt.xlabel("Time (hours)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()


```
