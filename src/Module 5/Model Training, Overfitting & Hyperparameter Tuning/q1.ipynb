Question 1: Understanding Overfitting with Polynomial Regression<br>
Task: Compare linear vs polynomial regression to observe overfitting.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline


np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**3 - X**2 + X + np.random.randn(*X.shape) * 2 

linear_model = LinearRegression()
linear_model.fit(X, y)

degree = 10
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_model.fit(X, y)

X_plot = np.linspace(-3, 3, 100).reshape(-1, 1)
y_linear_pred = linear_model.predict(X_plot)
y_poly_pred = poly_model.predict(X_plot)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Data')
plt.plot(X_plot, y_linear_pred, label='Linear Regression', color='blue')
plt.plot(X_plot, y_poly_pred, label=f'Polynomial Regression (degree {degree})', color='red')
plt.title("Linear vs Polynomial Regression (Overfitting Example)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
