```python
# Question 8: Parameter Tuning in DBSCAN
# Description: Experiment with different values of eps and min_samples in DBSCAN and observe the effect on clustering results.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

X, _ = make_moons(n_samples=300, noise=0.1, random_state=0)
X_scaled = StandardScaler().fit_transform(X)

eps_values = [0.1, 0.2, 0.3, 0.5]
min_samples_values = [3, 5, 10]

plt.figure(figsize=(15, 12))

for i, eps in enumerate(eps_values):
    for j, min_samples in enumerate(min_samples_values):

        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        y_dbscan = dbscan.fit_predict(X_scaled)
        
        plt.subplot(len(eps_values), len(min_samples_values), i * len(min_samples_values) + j + 1)
        plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_dbscan, cmap='viridis', s=50)
        plt.title(f"eps={eps}, min_samples={min_samples}")
        plt.xlabel("Feature 1")
        plt.ylabel("Feature 2")
        plt.grid(True)

plt.tight_layout()
plt.show()
```
