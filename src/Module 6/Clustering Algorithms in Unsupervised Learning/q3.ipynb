```python
# Question 2: Implementing Hierarchical Clustering
# Description: Use Hierarchical Clustering on a dataset and visualize the dendrogram.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

X, y = make_blobs(n_samples=150, centers=3, cluster_std=0.60, random_state=42)

linked = linkage(X, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,       
           color_threshold=15)         
plt.axhline(y=15, color='r', linestyle='--', label='Cutoff Threshold')  
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.legend()
plt.show()

cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_pred = cluster.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', s=50)
plt.title("Hierarchical Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()


```
